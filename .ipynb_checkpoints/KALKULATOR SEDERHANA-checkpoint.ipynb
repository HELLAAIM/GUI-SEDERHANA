```python
# import library
import pandas as pd
import numpy as np
from tkinter import *
import tkinter as tk
```

```python
#OOP

class App(tk.Tk):
    def __init__(window):
        super().__init__()
        
        #window.geometry("250x100")
        window.title("Login")
        window.resizable(1,1)
        
        #set grid
        window.columnconfigure(0, weight = 0)
        window.columnconfigure(1, weight = 0)
        
        expression =""
        window.create_widgets()
        
    def input_number(number, isi):
        global expression
        expression = expression +str(number)
        isi.set(expression)
        
    def evaluate(isi):
        global expression
        try:
            result = str(eval(expression))
            isi.set(result)
            expression = ""
        except:
                expression = ""
        
        
        
    def create_widgets(window):
        #nama
        isi = StringVar()
        nama_label = tk.Entry(window, text = isi)
        nama_label.grid(column = 0, row = 0, columnspan=4)
        
        isi.set("0")

        #button 123
        button_label = tk.Button(window, text = "1", bd = 2, command = lambda: input_number(2, isi))
        button_label.grid(column = 0, row = 1, sticky = tk.W, padx = 5, pady = 5)
        button_label.config(height = 1, width = 5)
        button_label = tk.Button(window, text = "2")
        button_label.grid(column = 1, row = 1, sticky = tk.W, padx = 5, pady = 5)
        button_label.config(height = 1, width = 5)
        button_label = tk.Button(window, text = "3")
        button_label.grid(column = 2, row = 1, sticky = tk.W, padx = 5, pady = 5)
        button_label.config(height = 1, width = 5)
        button_label = tk.Button(window, text = "+")
        button_label.grid(column = 3, row = 1, sticky = tk.W, padx = 5, pady = 5)
        button_label.config(height = 1, width = 5)
        
        #button 456
        button_label = tk.Button(window, text = "4")
        button_label.grid(column = 0, row = 2, sticky = tk.W, padx = 5, pady = 5)
        button_label.config(height = 1, width = 5)
        button_label = tk.Button(window, text = "5")
        button_label.grid(column = 1, row = 2, sticky = tk.W, padx = 5, pady = 5)
        button_label.config(height = 1, width = 5)
        button_label = tk.Button(window, text = "6")
        button_label.grid(column = 2, row = 2, sticky = tk.W, padx = 5, pady = 5)
        button_label.config(height = 1, width = 5)
        button_label = tk.Button(window, text = "-")
        button_label.grid(column = 3, row = 2, sticky = tk.W, padx = 5, pady = 5)
        button_label.config(height = 1, width = 5)
        
        #button 789
        button_label = tk.Button(window, text = "7")
        button_label.grid(column = 0, row = 3, sticky = tk.W, padx = 5, pady = 5)
        button_label.config(height = 1, width = 5)
        button_label = tk.Button(window, text = "8")
        button_label.grid(column = 1, row = 3, sticky = tk.W, padx = 5, pady = 5)
        button_label.config(height = 1, width = 5)
        button_label = tk.Button(window, text = "9")
        button_label.grid(column = 2, row = 3, sticky = tk.W, padx = 5, pady = 5)
        button_label.config(height = 1, width = 5)
        button_label = tk.Button(window, text = "=")
        button_label.grid(column = 3, row = 3, sticky = tk.W, padx = 5, pady = 5)
        button_label.config(height = 1, width = 5)

if __name__ == "__main__":
    app = App()
    app.mainloop()
        
```
